```python
# Start Ollama server as a background process and wait until it's ready
import subprocess, time, json, os, sys

def start_ollama_server(timeout_sec=30):
    # Start server
    server = subprocess.Popen(
        ["ollama", "serve"],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL
    )
    # Probe health endpoint until ready or timeout
    start = time.time()
    last_err = None
    while time.time() - start < timeout_sec:
        try:
            out = subprocess.run(
                ["curl", "-sS", "http://127.0.0.1:11434/api/tags"],
                capture_output=True, text=True, timeout=3
            )
            if out.returncode == 0:
                # Try to parse JSON to ensure server is truly up
                _ = json.loads(out.stdout or "[]")
                return True
        except Exception as e:
            last_err = e
        time.sleep(1)
    if last_err:
        print(f"Last error: {last_err}", file=sys.stderr)
    return False

ok = start_ollama_server(timeout_sec=45)
print("Ollama server ready:", ok)
```
